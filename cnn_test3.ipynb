```
directory = '/content/gdrive/My Drive/'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization, LeakyReLU, Input, Dot, merge, Permute, Reshape, Activation, Conv1D, MaxPooling1D
from keras import optimizers, activations
from keras.callbacks import ModelCheckpoint
from keras.backend import transpose
from keras import backend as K
```

```
def make_model2():
    x = Input((226, 1))

    h = Conv1D(filters=16, kernel_size=3, padding='valid')(x)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)

    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=16, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = MaxPooling1D(pool_size=2)(h)

    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)

    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=32, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = MaxPooling1D(pool_size=2)(h)

    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)

    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=64, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = MaxPooling1D(pool_size=2)(h)


    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)

    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=128, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = MaxPooling1D(pool_size=2)(h)

    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)
    h = Activation(swish)(h)
    h = Conv1D(filters=256, kernel_size=3, padding='same')(h)

    h = MaxPooling1D(pool_size=2)(h)

    h = Flatten()(h)

    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    
    h = BatchNormalization()(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)
    h = Activation(swish)(h)
    h = Dense(units=226, kernel_initializer='he_uniform')(h)

    h = Dense(units=4, kernel_initializer='he_uniform', activation='linear')(h)
    
    model = Model(input=x, output=h)
    return model

    
def swish(x) :
    return x * activations.sigmoid(x)
```

```
model = make_model2()
model.compile(loss='mae', optimizer=optimizers.Adam(0.0003), metrics=['mae'])
model.summary()
label_set = df[df.columns[:4]]
train_set = df[df.columns[4:]]
train_set = np.expand_dims(train_set, axis=2)
filepath= directory + "cnn_deep_model.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='mse', verbose=1, save_best_only=True, mode='auto')

model.fit(train_set, label_set, epochs=200, batch_size=1000, callbacks=[checkpoint], validation_split=0.05)

test = pd.read_csv(directory + 'test.csv', index_col=0)
test_X = test.values
test_X = np.expand_dims(test_X, axis=2)
print(test_X.shape)
pred_test = model.predict(test_X)
sample_sub = pd.read_csv(directory + 'sample_submission.csv', index_col=0)
submission = sample_sub+pred_test
submission.to_csv(directory + 'submission.csv')
```
